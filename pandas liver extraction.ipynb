## multiple condition

```python
# by using and(&)logical operator 

filter2=df[(df["LiverPatient"]==1) & (df["Age"]>=70)]
filter2['LiverPatient'].value_counts()
```

```python
# by using or(|)logical operator 

filter1=df[(df["Age"]==65) | (df["Gender"]=="Female")]
filter1['Gender'].value_counts()
```

```python
filter3=df[(df["Gender"]=="Female") | (df.Age>=35) | (df.DB<=6)]
filter3
```
