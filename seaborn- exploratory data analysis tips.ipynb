# plots for continous data

1- **univariate-(single variable at a time)**
- histrogram
- box plot

2- **bivariate-( two variable at a time)**

- scatter plot
- line plot
- violin plot
- join plot

3- **multi variate-(more than two variable at a time)**

- heat map
- pair plot

# histrogram / distribution plot

- it is used the check the given data is normal distribution or not

```python
sns.displot(df['total_bill'])
plt.show()
```

```python
sns.displot(df.tip)
plt.show()
```

```python
sns.displot(df['total_bill'],kde=True, bins=15) # kde-means kernal density estimation  
plt.title('histrogram od total bill')
plt.show()
```

## to check skewed distribution

```python
df['total_bill'].skew() # this is a right skewed distribution
```

```python
df['tip'].skew()
```

# pair plot ( multiple continuous variable)

```python
df.head()
```

# heat map
- it provide the relations between the variables

- it is very important in feature selection

```python
cor_matrix= df[['total_bill','tip']].corr()
cor_matrix
```

```python
corr=df.corr()
corr
```

```python
cor_matrix= df[continuous].corr()
cor_matrix
```

```python
sns.heatmap(cor_matrix,annot=True)
plt.show()
```

as per plot
- darker color gives lower correlation
- lighter color give higher correlation
