# Pandas series

 **1. Creating series using list**

```python
l=['banana',50]
s = pd.Series(l)
print(s)
```

```python
s=pd.Series(l,index=['a','b'])
print(s)
```

**2. create series using array**

```python
a=np.array(l)
print(a)
```

```python
s=pd.Series(a, index=['c','d'])
print(s)
```

**3. creating series using Dictonary**

```python
d={'a':1,'b':2,'c':3}
e = pd.Series(d)
print(e)
```

```python
d={'a':1,'b':2,'c':3}
e = pd.Series(d,index=['a','z','c'])
print(e)
```

```python
d={'a':1,'b':2,'c':3}
e = pd.Series(d,index=['a','c'])
print(e)
```

**in series dictonary used for to extract the data and list , array are use to assign the value**

```python
d=[1,2,5,6]
pd.Series(d,index=['a','b','c',"d"])
```

```python
a={'col1':[1,2],'col2':[3,4]}
b=pd.Series(a)
b
```

# merging of DataFrame

```python
a=pd.merge(t,h,on= 'city',how='inner')
a
```
