```python
days=np.arange(1,8)
max_t=np.arange(40,75,5)
min_t=[38,35,37,42,41,30,25]
avg_t=[46,52,54,62,68,43,58]

plt.plot(days,max_t,label='max', linewidth=2)
plt.plot(days,min_t, label='min', linewidth=3)
plt.plot(days,avg_t, label='avg', linewidth=4)

# set axis name and chart title

plt.xlabel('days')
plt.ylabel('Temperature')
plt.title("Weather")
plt.legend()
plt.show()
```

# Subplot

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]

plt.subplot(3,2,1)
plt.plot(x,y, c='red',marker='*', linestyle='dashed')
plt.title('Line plot')

plt.subplot(3,2,2)
plt.scatter(x,y)
plt.title('Scatter plot')

plt.subplot(3,2,3)
plt.plot(x,y, linewidth=3)

plt.subplot(3,2,6)
plt.plot(x,y,c='green',marker='o', linestyle='dashed')
plt.show()
```

# Pie chart

```python
list=[20,30,15,10,20,5]

lbs=['facecream','facewash','toothpaste','soap','sampoo','moisture ']

plt.pie(list, labels=lbs, autopct='%1.1f%%',radius=1.5, explode=(0,0.3,0,0,0,0.2))

plt.show()
```

```python
# counter clock and angle properties
plt.pie(list, labels=lbs, autopct='%1.1f%%',explode=(0,0.1,0,0,0,0),counterclock='False',startangle=90)
plt.axis('equal')

plt.show()
```

## data to plot

```python
la=['python','c++','jave','ruby']
sizes=[35,30,20,15]
color=['gold','blue','pink','purple']
plt.pie(sizes,labels=la,colors=color,radius=1.5,autopct='%1.1f%%')
plt.show()
```

```python
la=['deep','panda','jany','lado']
sizes=[50,35,15,40]
color=['gold','blue','green','red']
explode=(0.1,0,0.1,0)
plt.pie(sizes,labels=la,explode=explode,colors=color,radius=1.5,autopct='%1.1f%%')
plt.savefig('piechart.jpg')
plt.show()
```

# Bar Plot

- **it describe the comparison  between the discrete categorical data**
- **it can be ploted both horizontally and vertically**

```python
company=['google','amazon','myntra','fb']
revenue=[90,136,85,27]
plt.bar(company,revenue)
plt.ylabel('revenue $')
plt.title('us technology stock')
plt.show()
```

# Stack bar plot

```python
company=['google','amazon','myntra','fb']
revenue=[90,136,85,27]
profit=[40,20,34,12]
plt.bar(company,revenue, label='revenue')
plt.bar(company,profit, label='profit')
plt.title('us technology stock')
plt.legend()
plt.show()
```

# ploting bar graph seperately in both revenue and profit in one graph 

```python
company=['google','amazon','myntra','fb']
revenue=[90,136,85,27]
profit=[40,20,34,12]
xpos=np.arange(len(company))

plt.bar(xpos-0.2,revenue,width=0.4,label='revenue')
plt.bar(xpos+.2,profit,width=0.4,label='profit')

plt.xticks(xpos,company) # used to specify the company bar name
plt.ylabel('revenue')

plt.title('us technology stock')
plt.legend()
plt.show()
```

# creating different bar-chart in one images 

```python
company=['myntra','flipkart','amazon','alibaba','snapdeal']

plt.figure(1,figsize=(15,15))
q1_profit=[80,120,50,68,90]
q2_profit=[38,40,105,65,80]
q3_profit=[40,80,75,55,110]
q4_profit=[75,61,90,105,115]

plt.subplot(3,2,1)
plt.bar(company,q1_profit)
plt.title('quarter1 profit')

plt.subplot(3,2,2)
plt.pie(q2_profit,labels=company,autopct='%1.1f%%')
plt.title('quarter2 profit')


plt.subplot(3,2,3)
plt.pie(q3_profit,labels=company,autopct='%1.1f%%',radius=1,explode=(0,0.1,0,0,0) )
plt.title('quarter3 profit')

plt.subplot(3,2,5)
plt.bar(company,q3_profit)
plt.title('quarter3 profit')

plt.subplot(3,2,4)
plt.bar(company,q4_profit)
plt.title('quarter4 profit')
plt.savefig('subplot.jpg')
plt.show()
```

```python
company=['myntra','flipkart','amazon','alibaba','snapdeal']

plt.figure(1,figsize=(10,15))

q1_profit=[80,120,50,68,90]
q2_profit=[38,40,105,65,80]
q3_profit=[40,80,75,55,110]
q4_profit=[75,61,90,105,115]

plt.subplot(2,2,1)
plt.bar(company,q1_profit)
plt.title('quarter1 profit')

plt.subplot(2,2,2)
plt.bar(company,q2_profit)
plt.title('quarter2 profit')

plt.subplot(2,2,3)
plt.bar(company,q3_profit)
plt.title('quarter3 profit')

plt.subplot(2,2,4)
plt.bar(company,q4_profit)
plt.title('quarter4 profit')

plt.suptitle('profit on quarterly basis')
plt.show()


```
